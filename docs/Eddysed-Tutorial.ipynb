```python
df = pd.DataFrame(index=[ i for i in range(opd.shape[0]*opd.shape[1])], columns=['lvl','w','opd','w0','g0'])
i = 0 
LVL = []
WV,OPD,WW0,GG0 =[],[],[],[]
for j in range(opd.shape[0]):
       for w in range(opd.shape[1]):
            LVL +=[j+1]
            WV+=[w+1]
            OPD+=[opd[j,w]]
            WW0+=[w0[j,w]]
            GG0+=[g0[j,w]]
df.iloc[:,0 ] = LVL
df.iloc[:,1 ] = WV
df.iloc[:,2 ] = OPD
df.iloc[:,3 ] = WW0
df.iloc[:,4 ] = GG0
```
